# Storing Python results for import in Typst

```python
from IPython.display import display, JSON
import sympy as sp
```

## Simple results

Each of the following cells produces a cell result:

```python
#| label: some-number
2+2
```

```python
#| label: some-text
'some text'
```

The following cell produces a stream but no result (a cell can output results, streams, display items and errors):

```python
#| label: some-stream
print('some text')
```

## Multiple outputs

The following cell produces two display outputs (and no result):

```python
#| label: several-outputs
display(2+2)
display('some text')
```

## Generated math formula

```python
#| label: some-math
x, i, n = sp.symbols('x,i,n')
sp.Sum(x**i, (i, 1, n))
```

## JSON outputs

```python
some_dict = {
  'a': 0,
  's': 'some string',
}
```

```python
#| label: json-result

JSON(some_dict)
```

```python
#| label: several-json-outputs

for i in range(5):
    display(JSON({chr(97+i): i, 's': 'some string'}))
```
